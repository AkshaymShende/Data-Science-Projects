```
tf_idf = TfidfVectorizer(stop_words='english')

data_df['description']= data_df['description'].fillna('')

tf_matrix = tf_idf.fit_transform(data_df.description)

tf_matrix.shape
```

```
a= tf_idf.get_feature_names()
len(a)
```

```
cosine_sim = linear_kernel(tf_matrix,tf_matrix)

cosine_sim.shape
```

```
cosine_sim[0]
```

```
#Construct a reverse map of indices and movie titles

title_series = pd.Series(data_df.index, index=data_df.title).drop_duplicates()

title_series
```

```
def recommender(movie_title):

  movie_index = title_series[movie_title]

  sim_scores = list(enumerate(cosine_sim[movie_index]))

  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  similar_movie_index = [i[0] for i in sim_scores]

  similar_movie_names = [data_df['title'][i] for i in similar_movie_index]

  s = [':   '.join(x) for x in zip(data_df['title'], data_df['type'])]

  return s[:5]
```

```
recommender('Transformers Prime')
```
